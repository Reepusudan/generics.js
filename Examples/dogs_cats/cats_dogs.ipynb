**Retaining from a pretrained model**

```
var fs = require('fs');
var request = require('request');

var download = function(uri, filename, callback){
  request.head(uri, function(err, res, body){
    request(uri).pipe(fs.createWriteStream(filename)).on('close', callback);
  });
};


require('child_process').execSync('unzip /content/generics.js/Examples/dogs_cats/trained.zip');

var net = null;
var util = new gen.Utilities();
var dir2 = "/content/dogs_cats/dogs_cats.json";
util.restore_model(dir2).then(function(net2){
     net=net2;
});
```

**Predicting from a pretrained model**

```
var conv_kernel = [[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]];
var img_length=100;
var img_height=100;
var kernel_size=2;
var conv_options = "direct";
var img_process = new gen.Convolution();
var keys = ["cats","dogs"];
function get_index(value) {
  if (value > 0.49) {
    return 1;
  } else {
    return 0;
  }
}
download('https://www.statnews.com/wp-content/uploads/2019/08/AdobeStock_182445295-768x432.jpeg', 'dog1.jpeg', function () {
    var img_dir = "/content/dog1.jpeg";
    img_process.image_flatten(img_dir, img_length, img_height, kernel_size, conv_options, conv_kernel).then(function (mat) {
      console.log("I predicted it as a --> "+keys[get_index(util.predict(net,mat[0]))]);
    });
});
```
